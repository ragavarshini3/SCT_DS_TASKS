```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/drive/MyDrive/SCT_TASK/accident dataset task_4.csv")
df.head(12)
```

```python
print("Dataset Overview:\n", df.head())
print("\nColumn Information:\n", df.info())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df = df.dropna(subset=['Road_surface_type'])
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Lanes_or_Medians', data=df, palette='viridis')
plt.title("Accident Count by Lanes or Medians")
plt.xlabel("Lanes or Medians")
plt.ylabel("Accident Count")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(y='Weather_conditions', data=df, order=df['Weather_conditions'].value_counts().index, palette='coolwarm')
plt.title("Accidents by Weather Conditions")
plt.xlabel("Count")
plt.ylabel("Weather Condition")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Light_conditions', y='Type_of_collision', data=df, alpha=0.5)
plt.title("Accident Hotspots")
plt.xlabel("Light_conditions")
plt.ylabel("Type_of_collision")
plt.show()
```

```python
numerical_features = df.select_dtypes(include=['number']).columns
numerical_df = df[numerical_features]

plt.figure(figsize=(10, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap='viridis', linewidths=0.5) # Changed cmap to a valid option
plt.title("Correlation Between Factors")
plt.show()
```

```python
if 'day_of_week' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(x='Sex_of_driver', data=df, order=['Male','Female'], palette='magma')
    plt.title("Accidents by Day of the Week")
    plt.xlabel("Educational_level")
    plt.ylabel("Vehicle_driver_relation	")
    plt.show()
else:
    print("Column 'day_of_week' not found in the DataFrame.")
```

```python
if 'Accident_severity' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(x='Accident_severity', data=df, palette='coolwarm')
    plt.title("Accidents by Severity")
    plt.xlabel("Severity Level")
    plt.ylabel("Number of Accidents")
    plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Sex_of_driver'].value_counts()
    for x_label, grp in df.groupby('Age_band_of_driver')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Age_band_of_driver')
_ = plt.ylabel('Sex_of_driver')
```
